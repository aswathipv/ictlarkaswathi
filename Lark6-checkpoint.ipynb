# Read dataset

```python
import pandas as pd
data=pd.read_excel('Dataset_w6.xlsx')
```

# 1.Count of countries

```python
data['country'].count()
```

# 2. Data visualisations

```python
import matplotlib.pyplot as plt
datag=data.nlargest(5,['GDP per capita (current US$)'])
plt.plot(datag['country'],datag['GDP per capita (current US$)'])
```

```python
import seaborn as sns
dataa=data.nlargest(4,['Population in thousands (2017)'])
sns.barplot(dataa['country'],dataa['Population in thousands (2017)'])
```

```python
datas=data.nlargest(4,['Surface area (km2)'])
sns.barplot(datas['country'],datas['Surface area (km2)'])
```

# 3. Count of countries with respect to Regions 

```python
data.groupby('Region')[['country']].count()
```

# 4a.Highest GDP

```python
datagd=data.groupby('Region')[['GDP per capita (current US$)']].mean()
```

```python
datagd.sort_values('GDP per capita (current US$)',ascending=False)
```

# 4b. Highest International trade

```python
datait=data.groupby('Region')[['International trade: Imports (million US$)']].mean()
```

```python
datait.sort_values('International trade: Imports (million US$)',ascending=False)
```

# 5.Consistent wrt GDP

```python
datagds=data.groupby('Region')[['GDP per capita (current US$)']].std()
```

```python
datagds.sort_values('GDP per capita (current US$)')
```
