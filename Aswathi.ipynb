```python
plt.scatter(data[data['Classification']=='Iris-setosa']['SW'],data[data['Classification']=='Iris-setosa']['SL'])
plt.scatter(data[data['Classification']=='Iris-versicolor']['SW'],data[data['Classification']=='Iris-versicolor']['SL'])
plt.scatter(data[data['Classification']=='Iris-virginica']['SW'],data[data['Classification']=='Iris-virginica']['SL'])
plt.title("Scatter plot of Sepal width and Sepal length",fontsize=10)
plt.xlabel("Sepel Width",fontsize=10)
plt.ylabel("Sepel length",fontsize=10)
```
